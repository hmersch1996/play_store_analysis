```python
from google_play_scraper import app, Sort, reviews_all
from app_store_scraper import AppStore
import pandas as pd
import numpy as np
import json, os, uuid
```

```python
# en el Google Play Store el link de la app a analizar es:
# https://play.google.com/store/apps/details?id=py.com.elcomercio.retailbanking&hl=es_PY
google_app_id = 'py.com.elcomercio.retailbanking'

# En la Apple Store, el link de la misma app es:
# https://apps.apple.com/py/app/ueno/id1592509395
apple_app_name = 'ueno'
apple_app_id = '1592509395'
```

```python
# Extraccion de datos de la Google Play Store

from google_play_scraper import app, reviews_all

#app_id = 'py.com.elcomercio.retailbanking'
result = app(google_app_id)

reviews = reviews_all(
    google_app_id,
    sleep_milliseconds=0,  # Tiempo de espera entre solicitudes (opcional)
    lang='es',  # Idioma de los reviews
    country='py'  # Pa√≠s
)

print(f"Nombre de la aplicaci√≥n: {result['title']}")
print(f"Total de reviews: {len(reviews)}")

```

```python
df1 = pd.DataFrame(reviews)

df1
```

```python
# Extraccion de datos de la Apple App Store

from app_store_scraper import AppStore

app_ueno = AppStore(country="py", app_name=apple_app_name, app_id=apple_app_id)
app_ueno.review(how_many=1000)  # N√∫mero de reviews a obtener (puedes ajustar este n√∫mero)

reviews = app_ueno.reviews

print(f"Nombre de la aplicaci√≥n: {app_ueno.app_name}")
print(f"Total de reviews: {len(reviews)}")

```

```python
df2 = pd.DataFrame(reviews)

df2
```

```python
df1.to_csv("Data/google_play.csv",header=None)

df2.to_csv("Data/apple_store.csv",header=None)
```
